```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)
x = ['Question 8', 'Question 9', 'Question 10']
yes = [23,22,22]
no = [7,8,8]
x_axis = np.arange(len(x))
plt.bar(x_axis - 0.2, yes, 0.4, label = 'YES')
plt.bar(x_axis + 0.2, no, 0.4, label = 'NO')
plt.xticks(x_axis,x)
plt.xlabel("Questions")
plt.ylabel("Responses")
plt.title("CONSUMER AWARENESS")
plt.legend()
plt.show()
```

```python
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)
x = ['Question 11']
x_axis = np.arange(1)
same = [18]
less = [8]
more = [4]
plt.bar(x_axis - 0.2, same, 0.2, label = 'SAME')
plt.bar(x_axis, less, 0.2, label = 'LESS')
plt.bar(x_axis + 0.2, more, 0.2, label = 'MORE')
plt.ylim(0,25)
plt.xticks(x_axis,x)
plt.ylabel("Responses")
plt.title("CONSUMER AWARENESS")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
responses = ['YES','NO']
data = [13,17]
fig = plt.figure(figsize =(8, 6), dpi = 80)
plt.pie(data, labels = responses)
plt.legend(title = "RESPONSES for Question 12:", loc="upper left")
plt.show()
```
